```python
# Importing all the necessary libraries
import numpy as np 
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
import os

from keras.models import Sequential
from keras.layers import Conv2D, Dropout, MaxPooling2D, Dense, Flatten
```

```python
fresh = []
halfFresh = []
spoiled = []
folderName = "/kaggle/input/meat-freshness-image-dataset/Meat Freshness.v1-new-dataset.multiclass/train"
for filename in os.listdir('/kaggle/input/meat-freshness-image-dataset/Meat Freshness.v1-new-dataset.multiclass/train'):
    fileName = os.path.join(folderName, filename)
    if (filename.startswith("FRESH") and filename.endswith(".jpg")):
        fresh.append(fileName)
    elif (filename.startswith("SPOILED") and filename.endswith(".jpg")):
        spoiled.append(fileName)
    elif(filename.startswith("HALF") and filename.endswith(".jpg")):
        halfFresh.append(fileName)
        
        
data = []
labels = []

for label, image_list in enumerate([fresh, halfFresh, spoiled]): # 0 for fresh, 1 for halfFresh and 2 for spolied
    for image_path in image_list:
        image = plt.imread(image_path)
        data.append(image)
        labels.append(label)
```

```python
data = np.array(data)
labels = np.array(labels)
print(data.shape, labels.shape)
```

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(416,416,3)),
    MaxPooling2D((2,2)),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(filters=128, kernel_size=(3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(filters=256, kernel_size=(3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(filters=256, kernel_size=(3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(filters=512, kernel_size=(3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Dropout((0.4)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout((0.4)),
    Dense(3, activation='softmax')
    
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(data, labels, epochs=10)
```

```python
model.fit(data, labels, epochs=10)
```

```python
validFresh, validSpoiled, validHalfFresh = [], [], []

folderName = "/kaggle/input/meat-freshness-image-dataset/Meat Freshness.v1-new-dataset.multiclass/valid"
for filename in os.listdir(folderName):
    fileName = os.path.join(folderName, filename)
    if (filename.startswith("FRESH") and filename.endswith(".jpg")):
        validFresh.append(fileName)
    elif (filename.startswith("SPOILED") and filename.endswith(".jpg")):
        validSpoiled.append(fileName)
    elif(filename.startswith("HALF") and filename.endswith(".jpg")):
        validHalfFresh.append(fileName)
        
        
validData = []
validLabels = []

for label, image_list in enumerate([validFresh, validHalfFresh, validSpoiled]): # 0 for fresh, 1 for halfFresh and 2 for spolied
    for image_path in image_list:
        image = plt.imread(image_path)
        validData.append(image)
        validLabels.append(label)
        
validData = np.array(validData)
validLabels = np.array(validLabels)
```

```python
testLoss, testAccuracy = model.evaluate(validData, validLabels)
print(f'Test Accuracy: {testAccuracy}')
```
